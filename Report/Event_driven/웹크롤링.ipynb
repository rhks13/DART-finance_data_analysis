```python
#라이브러리 import하기
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#한 페이지만 가져오기
#HTML 코드 받아오기
url= 'https://www.law.go.kr/LSW/nwRvsLsPop.do?cptOfi=1480000'
html = requests.get(url)
html.text
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
soup
```

```python
tags = soup.select(".tbl3 tr")
tags
```

```python
tags[1].text
```

```python
web_list = []
for tag in tags:
  web_list.append(tag.text.strip().split('\n'))
```

```python
web_list
```

```python
main = pd.DataFrame(web_list)
main
```

```python
for i in range(2,92):
  new_url= f'https://www.law.go.kr/LSW/nwRvsLsPop.do?pg={i}&chrIdx=0&lsKndCd=&cptOfi=1480000&searchType=lsNm&lsNm=&p_spubdt=&p_epubdt=&p_spubno=&p_epubno='
  new_html = requests.get(new_url)
  new_soup = BeautifulSoup(new_html.text, 'html.parser')
  new_tags = new_soup.select(".tbl3 tr")
  new_web_list = []
  for tag in new_tags:
    new_web_list.append(tag.text.strip().split('\n'))
  df = pd.DataFrame(new_web_list)
  df = df.iloc[1:]
  main = pd.concat([main, df])
```

# **네이버 뉴스 크롤링**

```python
new_url= f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%B0%94%EC%9D%B4%EB%93%A0%20%EC%B9%9C%ED%99%98%EA%B2%BD%20%EC%A0%95%EC%B1%85&sort=0&photo=3&field=0&pd=0&ds=&de=&cluster_rank=22&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1'
new_html = requests.get(new_url)
new_soup = BeautifulSoup(new_html.text, 'html.parser')
```

```python
new_tags = new_soup.select(".news_tit")
new_tags
```

```python
new_web_list = []
for tag in new_tags:
  new_web_list.append(tag.text.strip().split('\n')[0])
new_web_list
```

```python
#sp_nws179 > div > div > div.news_info > div.info_group > span:nth-child(3)
```

```python
day = new_soup.select('div > div > div.news_info > div.info_group > span.info')
day
```

```python
day_list = []
for i in range(len(day)):
  if i%2 == 1 : 
    day_list.append(day[i].text)
day_list
```

```python
df = pd.DataFrame({'제목': new_web_list,'일시':day_list})
df
```

```python
ori = df.copy()
```

```python
for i in range(1,300):
  url= f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%B0%94%EC%9D%B4%EB%93%A0%20%EC%B9%9C%ED%99%98%EA%B2%BD%20%EC%A0%95%EC%B1%85&sort=0&photo=3&field=0&pd=0&ds=&de=&cluster_rank=11&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={i}1'
  html = requests.get(url)
  soup = BeautifulSoup(html.text, 'html.parser')
  tags = soup.select(".news_tit")
  web_list = []
  for tag in tags:
    web_list.append(tag.text.strip().split('\n')[0])
  day = soup.select('div > div > div.news_info > div.info_group > span.info')
  day_list = []
  for j in range(len(day)):
    if j%2 == 1 : 
      day_list.append(day[j].text)
  df1 = pd.DataFrame({'제목': web_list,'일시':day_list})
  ori = pd.concat([ori,df1])
```

```python
def find_key(text,word):
    count = 0
    start = 0
    word_len = len(word)
    while True:
        start = text.find(word, start)
        if start == -1:
            break
        if start > 0 and text[start-1].isalpha():
            start += 1
            continue
        if start + word_len < len(text) and text[start+word_len].isalpha():
            start += 1
            continue
        count += 1
        start += 1
    return count
```

```python
def token(text):
  
  return re.sub('[은는이가을를과의]$', "", text)
```

```python
from transformers import BertModel, BertTokenizer

model = BertModel.from_pretrained("monologg/kobert")
tokenizer = BertTokenizer.from_pretrained("monologg/kobert")
```

```python
from kobert_transformers import get_tokenizer

tokenizer = get_tokenizer()

text = "안녕하세요, 한국어 임베딩입니다."
tokens = tokenizer.tokenize(new_df.iloc[0,0])
print(tokens)
```

```python
input_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
import torch

input_ids = torch.tensor([input_ids])
attention_mask = torch.ones(input_ids.shape, dtype=torch.long, device=input_ids.device)

# 패딩을 수행할 위치를 attention_mask로 설정
# 패딩할 위치는 0, 아닌 위치는 1
for i, id in enumerate(input_ids[0]):
    if id == tokenizer.pad_token_id:
        attention_mask[0][i] = 0
```

```python
outputs = model(input_ids, attention_mask)
last_hidden_states = outputs[0]
print(last_hidden_states.shape)
```

```python
def embed(text):
  tokens = tokenizer.tokenize(text)
  input_ids = tokenizer.convert_tokens_to_ids(tokens)
  input_ids = torch.tensor([input_ids])
  attention_mask = torch.ones(input_ids.shape, dtype=torch.long, device=input_ids.device)
  for i, id in enumerate(input_ids[0]):
    if id == tokenizer.pad_token_id:
        attention_mask[0][i] = 0
  outputs = model(input_ids, attention_mask)
  last_hidden_states = outputs[0]
  return last_hidden_states
```

```python
new_df.applymap(embed)
```
